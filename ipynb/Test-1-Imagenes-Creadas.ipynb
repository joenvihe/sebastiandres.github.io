## 3 - Utilizando matplotlib

```python
%matplotlib inline
import pylab
x = pylab.linspace(-1.,1.,101)
y = x*(x-0.5)*(x+0.5)
pylab.figure()
pylab.plot(x,y,lw=2.0,label='y=y(x)')
pylab.xlabel('x')
pylab.ylabel('y')
pylab.title('Mi Ejemplo')
pylab.grid('on')
pylab.legend()
pylab.figure()
pylab.plot(x,y**2,lw=2.0,label='y=y(x)')
```

```python
import numpy as np
pylab.figure()
pylab.plot(x,np.sin(2*np.pi*x),"r", lw=2.0,label='y=y(x)')
```
